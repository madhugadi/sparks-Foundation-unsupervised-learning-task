```python
X = data.drop(['Id', 'Species'],axis = 1)
X = np.array(X)
X[:5] # first 5 rows
```

### Elbow curve
#### plot the graph of k(number of clusters) vs SSE
#### SSE is sum of squared distances of data points to their closest cluster center
#### choose appropriate k where error is minimum or significantly not reducing further

```python
from sklearn.cluster import KMeans ##Importing Kmeans clss from SKlearn
wcss = [] ##Within cluster sum of squares
for i in range(1, 11):
    model = KMeans(n_clusters = i,random_state = 0)
    model.fit(X)
    wcss.append(model.inertia_)
```

```python
plt.figure(figsize=(18,11))
plt.plot(range(1, 11), wcss, marker = 'o')
plt.title('The Elbow Curve', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show();
```

```python
model = KMeans(n_clusters = 3, random_state = 0)
pred = model.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=model.labels_,cmap='rainbow')

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:, 1], s = 200, c = 'red' , label = 'centeroid',
            marker='*')  ## 0 means x-axis and 1 means y-axis
```

### Alternative way of plotting clusters

```python
plt.scatter(X[pred == 0, 0],X[pred == 0, 1], s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'red', label = 'Iris-versicolour')
plt.scatter(X[pred == 2, 0], X[pred == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```
